```python
import torch

from torchvision.models.vision_transformer import vit_b_16

model = vit_b_16(weights="DEFAULT")
```

```python
device = torch.device("cpu")
model = model.to(device)
```

```python
image = torch.rand((1, 3, 224, 224), device=device)
```

```python
encoder_outputs = {}

def get_encoder_output(name):
    def hook(model, input, output):
        encoder_outputs[name] = output.detach()
    return hook

for layer in range(12):
    model.get_submodule(f"encoder.layers.encoder_layer_{layer}").register_forward_hook(get_encoder_output(f"encoder_layer_{layer}"))
```

```python
%%timeit
out = model(image)
```

```python
for k, v in encoder_outputs.items():
    print(k, v.shape)
```
